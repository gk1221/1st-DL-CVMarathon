## Day 019 作業：DataFrame 的排序及顯示

同 Day 017，請載入 2020/4/2 COVID-19 Daily Report 資料集，計算每個國家/區域的累計確診人數。

請依確診案例數，列出截至4月2日為止 (UTC 時間)，確診案例最多的10個國家/區域。

```julia
using DataFrames, CSV
```

```julia
df = CSV.read("04-02-2020.csv")
```

```julia
# 請在此輸入程式碼
s_df = sort(df, :Confirmed, rev=true)
```

```julia
s_df[! , [:Confirmed,:Country_Region]]
```

<h1>別名 : 欄位=> function => 新欄位</h1>

```julia
show(sort(combine(groupby(s_df, :Country_Region), :Confirmed=>sum=>:c_s), :c_s, rev=true), allrows=true)
```

```julia
show(sort(df, :Confirmed, rev=true)[1:10, [3, 4, 8, 9, 10, 11]])
```
